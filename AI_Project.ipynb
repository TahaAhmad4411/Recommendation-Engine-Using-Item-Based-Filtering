```python
import pandas as pd
import numpy as np
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_names)
df.head()
```

```python
movie_titles = pd.read_csv('movies.csv')
movie_titles= movie_titles.rename(columns={'movieId': 'item_id'})
movie_titles.head()
```

```python
df = pd.merge(df,movie_titles,on='item_id')
df.head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
ratings['num of ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

```python
US_user_ratings = moviemat['Usual Suspects, The (1995)']
US_user_ratings.head()
```

```python
similar_to_US = moviemat.corrwith(US_user_ratings)
```

```python
corr_US = pd.DataFrame(similar_to_US,columns=['Correlation'])
corr_US.dropna(inplace=True)
corr_US.head()
```

```python
corr_US.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_US = corr_US.join(ratings['num of ratings'])
corr_US.head()
```

```python
corr_US[corr_US['num of ratings']>50].sort_values('Correlation',ascending=False).head()
```
